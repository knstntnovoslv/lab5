```python
import pandas as pd
```

```python
data1 = pd.read_csv("Count_all_elements.csv")
data2 = pd.read_csv("Count_all_elements_in_group.csv")
data3 = pd.read_csv("Count_every_elements_in_group.csv")
data4 = pd.read_csv("Division_by_parts.csv")
data5 = pd.read_csv("subexpression_array_sequence.csv")
data6 = pd.read_csv("subexpression_string_array.csv")
data7 = pd.read_csv("subexpression_string_list.csv")
```

```python
dim = [k for k in range(10, 100, 10)]
dim
```

```python
data1['dimension'] = pd.Series(dim)
data2['dimension'] = pd.Series(dim)
data3['dimension'] = pd.Series(dim)
data4['dimension'] = pd.Series(dim)
data5['dimension'] = pd.Series(dim)
data6['dimension'] = pd.Series(dim)
data7['dimension'] = pd.Series(dim)
```

```python
res = data1.merge(data2,on='dimension', how='inner')
sheet = [data3, data4, data5, data6, data7]
for i in sheet:
    res = res.merge(i, on='dimension', how='inner')
```

```python
res
```

```python
res.loc[:,['count_all_elements', 'count_all_elements_in_group', 'division_by_parts', 'count_every_elements_in_group', 'dimension']].plot(kind='line', y=['count_all_elements', 'count_all_elements_in_group', 'division_by_parts', 'count_every_elements_in_group'], x='dimension', title='Hist visualisation')
```

```python
res.loc[:, ['subexpression_array_sequence', 'subexpression_string_array', 'subexpression_string_list', 'dimension']].plot(
    title="Subexpression visualisation", kind='line', y=['subexpression_array_sequence', 'subexpression_string_array', 'subexpression_string_list'], x='dimension'
)
```
